# Stacking

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, VotingClassifier

clf1 = LogisticRegression(random_state=1)
clf2 = RandomForestClassifier(random_state=1)
clf3 = GaussianNB()
```

```python
eclf1 = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3)], voting='hard')
eclf1 = eclf1.fit(trainingData, y)
eclf2 = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3)], voting='soft')
eclf2 = eclf2.fit(trainingData, y)
```

```python
eclf3 = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3)], voting='soft', weights=[2,2,1])
eclf3 = eclf3.fit(trainingData, y)
```

```python
eclf4 = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3), ('mlp', mlp)], voting='soft', weights=[2,2,1,2])
eclf4 = eclf4.fit(trainingData, y)
```

```python
for clf, label in zip([clf1, clf2, clf3, eclf1, eclf2, eclf3, eclf4], ['Logistic Regression', 'Random Forest', 'naive Bayes', 'Ensemble', 'Ensemble', 'Ensemble', 'Ensemble']):
    scores = cross_val_score(clf, trainingData, y, cv=10, scoring='accuracy')
    print("Training Data Accuracy: %0.2f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label))
```

```python
for clf, label in zip([clf1, clf2, clf3, eclf1, eclf2, eclf3, eclf4], ['Logistic Regression', 'Random Forest', 'naive Bayes','Ensemble', 'Ensemble', 'Ensemble','Ensemble']):
    scores = cross_val_score(clf, testData, testy, cv=10, scoring='accuracy')
    print("Test Data Accuracy: %0.2f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label))
```

# 가장 좋다고 생각되는 모델 (LR+RF+MLP ensemble)
